```python
#we are extracting only those which contains GSN no.
stations = {}

for line in open('stations.txt','r'):
    if 'GSN' in line:
        fields = line.split()
        
        stations[fields[0]] = ' '.join(fields[4:])
```

```python
#finding interesting pattern in station names
def findstation(s):
    found = {code: name for code,name in stations.items() if s in name}
    print(found)
```

```python
#ID has 11 columns, followed by year which has 4 columns, followed by months = 2, Element = 4, and then Value, Mflag, Qflag set is repititive for 31 lines.
#We are making a function parsefile which will parse the file according to our needs. 
def parsefile(filename):
    return np.genfromtxt(filename,
                         delimiter=dly_delimiter, 
                         usecols = dly_usecols,
                         dtype = dly_dtype,
                         names=dly_names)
#result will be a numpy record array
```

```python
dly_delimiter = [11,4,2,4]+[5,1,1,1]*31
#year = 1,Month = 2, Element = 3, and then Value'n' will be expressed as a multiple of 4 because we are only using value and not flag and every value column is multiple of four after ELEMENT.
dly_usecols = [1,2,3]+[4*n for n in range(1,32)]
dly_dtype = [np.int32, np.int32, (np.str_,4)]+[np.int32]*31
dly_names = ['year','month','obs']+[str(day) for day in range(1,31+1)]
```

```python
#the above data does not look good and comprehensive, we need to arrange it in a better way 
def better_rep(record):
    startdate = np.datetime64('{}-{:02}'.format(record['year'],record['month']))
    #for 1 month from startdate with a timestep of 1 day
    dates = np.arange(startdate,startdate+np.timedelta64(1,'M'),np.timedelta64(1,'D'))
    
    rows = [(date,record[str(i+1)]/10) for i,date in enumerate(dates)]
    
    return np.array(rows,dtype=[('date','M8[D]'),('value','d')])
```

```python
lihue = better_rep(lihue[0])
```

```python
lihue
```

```python
def getobs(filename,obs):
    return np.concatenate([better_rep(row) for row in parsefile(filename) if row[2]==obs])
```

```python
#ABOVE GRAPH IS LITTLE STRANGE, IT MUST BE -999.9 VALUE
def getobs(filename,obs):
    data = np.concatenate([better_rep(row) for row in parsefile(filename) if row[2]==obs])
    data['value'][data['value'] == -999.9] = np.nan
    return data
```

```python
lihue_tmax1 = getobs('USW00022536.dly','TMAX')
lihue_tmin1 = getobs('USW00022536.dly','TMIN')
```

```python
plt.plot(lihue_tmax1['date'],lihue_tmax1['value'])
plt.plot(lihue_tmin1['date'],lihue_tmin1['value'])
```

```python
#data is very noisy
#filling the missing value by interpolating the near values
#example of how numpy can interpolate

x = np.array([1,2,3,4],'d')
y = np.array([5,3,7,10],'d')

plt.plot(x,y,'o')
```

```python
#interpolation
xs = np.linspace(1,10)
ys = np.interp(xs,x,y)

plt.plot(xs,ys,'.')

plt.plot(x,y,'o')

plt.axis([0,5,0,12])
```

```python
def fillmissing(data):
    datesfloat = data['date'].astype(np.float64)
    nan = np.isnan(data['value'])
    
    data['value'][nan] = np.interp(datesfloat[nan], datesfloat[~nan],data['value'][~nan])
```

```python
fillmissing(lihue_tmax)
fillmissing(lihue_tmin)
```

```python
np.mean(lihue_tmax['value']),np.mean(lihue_tmin['value'])
```

```python
#smoothing out
def plot_smooth(t,win=10):
    smoothed = np.correlate(t['value'],np.ones(win)/win,'same')
    
    plt.plot(t['date'],smoothed)
```

```python
plot_smooth(lihue_tmax1[1000:1200])
```

```python
#comparison between the graphs

plt.plot(lihue_tmin1[1000:1200]['date'],lihue_tmin1[1000:1200]['value'])

plot_smooth(lihue_tmax1[1000:1200])
plot_smooth(lihue_tmax1[1000:1200],30)
```

```python
#for the datastations we computed above
plt.figure(figsize=(20,12))
for i,code in enumerate(datastations):
    plt.subplot(2,2,i+1)
    plot_smooth(getobs('{}.dly'.format(code),'TMIN'),365)
    plot_smooth(getobs('{}.dly'.format(code),'TMAX'),365)
    plt.title(stations[code])
    plt.axis(xmin=np.datetime64('1952'),xmax = np.datetime64('2012'),ymin = -10, ymax = 30)
```

```python
#computing daily records
##arranging all the temperatures in 2 dimensional array where each row represents year and each column represents day
#boolean masks
def selectyear(data,year):
    start = np.datetime64('{}'.format(year))
    end = start + np.timedelta64(1,'Y')
    return data[(data['date']>start)&(data['date']<end)]['value']
    
```
